```python
import mne
import os
from mne_bids import (
    BIDSPath, 
    write_raw_bids,
    write_meg_calibration,
    write_meg_crosstalk,
    write_anat
)
```

```python
# Prepare the meg raw data into BIDS format
raw_fname = "E:/MEGData/xuz/conjsemproj_20250926cy/250926/run1.fif"
emptyroom_fname = "E:/MEGData/xuz/250926/emptyroom.fif"

raw = mne.io.read_raw_fif(raw_fname, allow_maxshield=True)
raw_er = mne.io.read_raw_fif(emptyroom_fname, allow_maxshield=True)
bids_path = BIDSPath(subject="20250926cy", session="01", run="01", task="ConjSemProj",
                    datatype="meg", root="./data")
write_raw_bids(
    raw, 
    bids_path=bids_path, 
    empty_room=raw_er,
    overwrite=True
)

# Write calibration information
cal_fname = "E:/xuz/others/sss/sss_cal_3160_240307.dat"
write_meg_calibration(cal_fname, bids_path)

# Write crosstalk information
ct_fname = "E:/xuz/others/sss/ct_sparse_triux2.fif"
write_meg_crosstalk(ct_fname, bids_path)
```

```python
# Prepare the T1w data into BIDS format
t1_fname = os.path.join(
    "E:/xuz/SemSynProj/Formal-experiment_ver-3",
    "ConjSemProj_20250926ZXT_20250926114526/ConjSemProj_20250926ZXT",
    "MEG/MR_t1_gre_fsp3d_sag_iso0.7mm_ACS",
    "MR_t1_gre_fsp3d_sag_iso0.7mm_ACS_t1_gre_fsp3d_sag_iso0.7mm_ACS_20250926114526_201.nii"
)

t1w_bids_path = BIDSPath(subject="20250926zxt", task="ConjSemProj",
                         datatype="anat", root="./data", suffix="T1w")

write_anat(
    image=t1_fname,
    bids_path=t1w_bids_path,
    overwrite=True,
    verbose=True
)
```
